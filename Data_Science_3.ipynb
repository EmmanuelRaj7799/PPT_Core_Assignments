## Question 1 :
Scenario: A company wants to analyze the sales performance of its products in different regions. They have collected the following data:
   Region A: [10, 15, 12, 8, 14]

   Region B: [18, 20, 16, 22, 25]

   Calculate the mean sales for each region.

## Answer:
To calculate the mean sales for each region, you need to sum up the sales data for each region and divide it by the number of observations (data points) in that region.

For Region A:
Sales = [10, 15, 12, 8, 14]
Mean Sales (Region A) = (10 + 15 + 12 + 8 + 14) / 5 = 59 / 5 = 11.8

For Region B:
Sales = [18, 20, 16, 22, 25]
Mean Sales (Region B) = (18 + 20 + 16 + 22 + 25) / 5 = 101 / 5 = 20.2

# Therefore, the mean sales for Region A is 11.8 and for Region B is 20.2.

## Question 2
Scenario: A survey is conducted to measure customer satisfaction on a scale of 1 to 5. The data collected is as follows:

   [4, 5, 2, 3, 5, 4, 3, 2, 4, 5]

   Calculate the mode of the survey responses.

## Answer :
To calculate the mode of the survey responses, you need to determine the value(s) that appear most frequently in the data.

The given survey responses are:

[4, 5, 2, 3, 5, 4, 3, 2, 4, 5]

To find the mode, we can count the frequency of each value in the data and identify the value(s) with the highest frequency.

The frequency count for each value in the data is as follows:

2: 2 times

3: 2 times

4: 3 times

5: 3 times

From the frequency count, we can see that both 4 and 5 appear most frequently in the data, with a frequency of 3 times each. Therefore, the mode of the survey responses is 4 and 5.

It's important to note that the mode can be a single value or multiple values, depending on whether there is a tie for the highest frequency. In this case, both 4 and 5 have the highest frequency, so they are both considered the mode of the survey responses.

## Question 3 :

   Department A: [5000, 6000, 5500, 7000]

   Department B: [4500, 5500, 5800, 6000, 5200]

   Calculate the median salary for each department.

## Answer :
To calculate the median salary for each department, you need to arrange the salary data in ascending order and find the middle value(s) of the sorted data.

For Department A:

Salary data: [5000, 6000, 5500, 7000]

Arrange in ascending order: [5000, 5500, 6000, 7000]

Median Salary (Department A) = (5500 + 6000) / 2 = 11500 / 2 = 5750

For Department B:

Salary data: [4500, 5500, 5800, 6000, 5200]

Arrange in ascending order: [4500, 5200, 5500, 5800, 6000]

Median Salary (Department B) = 5500

Therefore, the median salary for Department A is 5750 and for Department B is 5500.

When the number of data points is odd, the median is the middle value. When the number of data points is even, the median is the average of the two middle values. In Department A, there are 4 data points, so the median is the average of the second and third values. In Department B, there are 5 data points, so the median is the third value.

## Question 4 :
Scenario: A data analyst wants to determine the variability in the daily stock prices of a company. The data collected is as follows:

   [25.5, 24.8, 26.1, 25.3, 24.9]

   Calculate the range of the stock prices.

## Answer :
To calculate the range of the stock prices, you need to find the difference between the highest and lowest values in the data.

The given stock prices are:

[25.5, 24.8, 26.1, 25.3, 24.9]

To find the range, we can identify the highest and lowest values in the data and calculate the difference between them.

Highest value: 26.1

Lowest value: 24.8

Range = Highest value - Lowest value

Range = 26.1 - 24.8

Range = 1.3

Therefore, the range of the stock prices is 1.3. The range provides a measure of the variability or spread of the stock prices by indicating the difference between the highest and lowest values.








## Question 5 :
Scenario: A study is conducted to compare the performance of two different teaching methods. The test scores of the students in each group are as follows:

   Group A: [85, 90, 92, 88, 91]

   Group B: [82, 88, 90, 86, 87]
   
   Perform a t-test to determine if there is a significant difference in the mean scores between the two groups.


```python
from scipy import stats

group_a = [85, 90, 92, 88, 91]
group_b = [82, 88, 90, 86, 87]

# Perform the t-test
t_stat, p_value = stats.ttest_ind(group_a, group_b, equal_var=False)

# Print the results
print("T-statistic:", t_stat)
print("p-value:", p_value)
```

## Question 7 :
Scenario: A survey is conducted to measure the heights of a group of people. The data collected is as follows:

   [160, 170, 165, 155, 175, 180, 170]
   
   Calculate the standard deviation of the heights.


```python
import numpy as np

heights = [160, 170, 165, 155, 175, 180, 170]

# Calculate the mean of the heights
mean_height = np.mean(heights)

# Calculate the squared differences from the mean
squared_diff = [(height - mean_height) ** 2 for height in heights]

# Calculate the average of the squared differences
mean_squared_diff = np.mean(squared_diff)

# Calculate the standard deviation
std_dev = np.sqrt(mean_squared_diff)

# Print the standard deviation
print("Standard Deviation:", std_dev)
```

## Question 9 :
Scenario: A study is conducted to compare the effectiveness of two different medications. The recovery times of the patients in each group are as follows:

   Medication A: [10, 12, 14, 11, 13]

   Medication B: [15, 17, 16, 14, 18]
   
   Perform an analysis of variance (ANOVA) to determine if there is a significant difference in the mean recovery times between the two medications.


```python
from scipy import stats

medication_a = [10, 12, 14, 11, 13]
medication_b = [15, 17, 16, 14, 18]

# Perform the one-way ANOVA test
f_stat, p_value = stats.f_oneway(medication_a, medication_b)

# Print the results
print("F-statistic:", f_stat)
print("p-value:", p_value)
```

[10.2, 9.8, 10.0, 10.5, 10.3, 10.1]

Perform a hypothesis test to determine if the mean weight differs significantly from 10 grams.


```python
from scipy import stats

weights = [10.2, 9.8, 10.0, 10.5, 10.3, 10.1]

# Perform the one-sample t-test
t_stat, p_value = stats.ttest_1samp(weights, 10)

# Print the results
print("T-statistic:", t_stat)
print("p-value:", p_value)
```

## Question 12 :
Scenario: A company wants to analyze the click-through rates of two different website designs. The number of clicks for each design is as follows:

  Design A: [100, 120, 110, 90, 95]

  Design B: [80, 85, 90, 95, 100]

  Perform a chi-square test to determine if there is a significant difference in the click-through rates between the two designs.


```python
from scipy import stats

design_a = [100, 120, 110, 90, 95]
design_b = [80, 85, 90, 95, 100]

# Perform the chi-square test of independence
chi2, p_value, _, _ = stats.chi2_contingency([design_a, design_b])

# Print the results
print("Chi-square statistic:", chi2)
print("p-value:", p_value)
```

## Question 13 :
  Scenario: A survey is conducted to measure customer satisfaction with a product on a scale of 1 to 10. The data collected is as follows:

  [7, 9, 6, 8, 10, 7, 8, 9, 7, 8]

  Calculate the 95% confidence interval for the population mean satisfaction score.


```python
import numpy as np
from scipy import stats

satisfaction_scores = [7, 9, 6, 8, 10, 7, 8, 9, 7, 8]

# Calculate the sample mean and standard deviation
sample_mean = np.mean(satisfaction_scores)
sample_std = np.std(satisfaction_scores, ddof=1)

# Calculate the t-value for the desired confidence level (95%)
confidence_level = 0.95
t_value = stats.t.ppf(1 - (1 - confidence_level) / 2, df=len(satisfaction_scores) - 1)

# Calculate the standard error
standard_error = sample_std / np.sqrt(len(satisfaction_scores))

# Calculate the confidence interval
lower_bound = sample_mean - t_value * standard_error
upper_bound = sample_mean + t_value * standard_error

# Print the confidence interval
print("Confidence Interval: [{:.2f}, {:.2f}]".format(lower_bound, upper_bound))
```

## Question 15 :
Scenario: A study is conducted to compare the preferences of two groups of participants. The preferences are measured on a Likert scale from 1 to 5. The data collected is as follows:

Group A: [4, 3, 5, 2, 4]

Group B: [3, 2, 4, 3, 3]

Perform a Mann-Whitney U test to determine if there is a significant difference in the median preferences between the two groups.


```python
from scipy import stats

group_a = [4, 3, 5, 2, 4]
group_b = [3, 2, 4, 3, 3]

# Perform the Mann-Whitney U test
statistic, p_value = stats.mannwhitneyu(group_a, group_b, alternative='two-sided')

# Print the results
print("Mann-Whitney U statistic:", statistic)
print("p-value:", p_value)
```

## Question 16:
Scenario: A company wants to analyze the distribution of customer ages. The data collected is as follows:

  [25, 30, 35, 40, 45, 50, 55, 60, 65, 70]
  
  Calculate the interquartile range (IQR) of the ages.


```python
import numpy as np

ages = [25, 30, 35, 40, 45, 50, 55, 60, 65, 70]

# Calculate the first quartile (Q1)
q1 = np.percentile(ages, 25)

# Calculate the third quartile (Q3)
q3 = np.percentile(ages, 75)

# Calculate the interquartile range (IQR)
iqr = q3 - q1

# Print the IQR
print("Interquartile Range (IQR):", iqr)
```

## Question 17 :
Scenario: A study is conducted to compare the performance of three different machine learning algorithms. The accuracy scores for each algorithm are as follows:

  Algorithm A: [0.85, 0.80, 0.82, 0.87, 0.83]

  Algorithm B: [0.78, 0.82, 0.84, 0.80, 0.79]

  Algorithm C: [0.90, 0.88, 0.89, 0.86, 0.87]

  Perform a Kruskal-Wallis test to determine if there is a significant difference in the median accuracy scores between the algorithms.


```python
from scipy import stats

algorithm_a = [0.85, 0.80, 0.82, 0.87, 0.83]
algorithm_b = [0.78, 0.82, 0.84, 0.80, 0.79]
algorithm_c = [0.90, 0.88, 0.89, 0.86, 0.87]

# Perform the Kruskal-Wallis test
statistic, p_value = stats.kruskal(algorithm_a, algorithm_b, algorithm_c)

# Print the results
print("Kruskal-Wallis statistic:", statistic)
print("p-value:", p_value)
```
