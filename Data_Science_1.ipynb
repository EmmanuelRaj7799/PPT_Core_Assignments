1. Write a Python program to reverse a string without using any built-in string reversal functions.

```python
def reverse_string(string):
    reversed_string = ""
    for i in range(len(string) - 1, -1, -1):
        reversed_string += string[i]
    return reversed_string

# Example usage
string = "Hello, World!"
reversed_string = reverse_string(string)
print(reversed_string)
```

13. Write a Python program to generate all permutations of a given string.

```python
def generate_permutations(string):
    # Base case: if the string has only one character
    if len(string) == 1:
        return [string]

    # List to store the permutations
    permutations = []

    # Iterate through each character in the string
    for i in range(len(string)):
        # Fix the current character at position i
        current_char = string[i]

        # Generate permutations of the remaining characters
        remaining_chars = string[:i] + string[i+1:]
        sub_permutations = generate_permutations(remaining_chars)

        # Append the current character to each permutation of the remaining characters
        for sub_permutation in sub_permutations:
            permutations.append(current_char + sub_permutation)

    return permutations

# Example usage
string = "ABC"
permutations = generate_permutations(string)
for permutation in permutations:
    print(permutation)
```

14. Implement a function to calculate the Fibonacci sequence up to a given number of terms.

```python
def fibonacci_sequence(n):
    sequence = []
    if n >= 1:
        sequence.append(0)
    if n >= 2:
        sequence.append(1)
    for i in range(2, n):
        next_term = sequence[i-1] + sequence[i-2]
        sequence.append(next_term)
    return sequence

# Example usage
terms = 10
fibonacci_sequence = fibonacci_sequence(terms)
print(fibonacci_sequence)
```

15. Write a program to find the median of a list of numbers.

```python
def find_median(numbers):
    sorted_numbers = sorted(numbers)
    n = len(sorted_numbers)
    if n % 2 == 0:
        # If the number of elements is even
        mid1 = n // 2
        mid2 = mid1 - 1
        median = (sorted_numbers[mid1] + sorted_numbers[mid2]) / 2
    else:
        # If the number of elements is odd
        mid = n // 2
        median = sorted_numbers[mid]
    return median

# Example usage
numbers = [4, 2, 6, 1, 5, 3]
median = find_median(numbers)
print(median)
```

```python
def max_subarray_sum(lst):
    max_sum = lst[0]
    current_sum = lst[0]

    for num in lst[1:]:
        current_sum = max(num, current_sum + num)
        max_sum = max(max_sum, current_sum)

    return max_sum

# Example usage
lst = [-2, 1, -3, 4, -1, 2, 1, -5, 4]
max_sum = max_subarray_sum(lst)
print(max_sum)
```

25. Write a Python program to merge two sorted lists into a single sorted list.

```python
def merge_sorted_lists(list1, list2):
    merged_list = []
    i = 0  # Index for list1
    j = 0  # Index for list2

    # Merge the two lists in sorted order
    while i < len(list1) and j < len(list2):
        if list1[i] <= list2[j]:
            merged_list.append(list1[i])
            i += 1
        else:
            merged_list.append(list2[j])
            j += 1

    # Append any remaining elements from list1
    while i < len(list1):
        merged_list.append(list1[i])
        i += 1

    # Append any remaining elements from list2
    while j < len(list2):
        merged_list.append(list2[j])
        j += 1

    return merged_list
```

```python
list1 = [1, 3, 5, 7]
list2 = [2, 4, 6, 8]
merged = merge_sorted_lists(list1, list2)
print(merged)
```

28. Implement a function to calculate the square root of a given number.

```python
def square_root(number):
    if number < 0:
        raise ValueError("Square root is undefined for negative numbers.")

    if number == 0:
        return 0

    # Initial guess
    guess = number / 2

    while True:
        new_guess = (guess + number / guess) / 2

        # Check if the guess has converged
        if abs(new_guess - guess) < 1e-9:
            return new_guess

        guess = new_guess
```

```python
number = 16
sqrt = square_root(number)
print(sqrt)  # Output: 4.0
```

30. Implement a function to find the minimum element in a rotated sorted list.

```python
def find_minimum(nums):
    left = 0
    right = len(nums) - 1

    # If the list is not rotated, return the first element
    if nums[left] < nums[right]:
        return nums[left]

    # Perform binary search to find the minimum element
    while left < right:
        mid = left + (right - left) // 2

        if nums[mid] > nums[right]:
            left = mid + 1
        else:
            right = mid

    return nums[left]
```

```python
nums = [4, 5, 6, 7, 0, 1, 2]
minimum = find_minimum(nums)
print(minimum)
```

34. Implement a function to find the longest common prefix among a list of strings.

```python
def longest_common_prefix(strings):
    if not strings:
        return ""

    prefix = strings[0]

    for string in strings[1:]:
        while string[:len(prefix)] != prefix:
            prefix = prefix[:-1]
            if not prefix:
                return ""

    return prefix
```

```python
strings = ["flower", "flow", "flight"]
prefix = longest_common_prefix(strings)
print(prefix)
```

42. Implement a function to find the longest palindrome substring in a given string.

```python
def longest_palindrome_substring(string):
    n = len(string)
    max_length = 0
    start = 0

    # Helper function to expand around center
    def expand_around_center(left, right):
        nonlocal start, max_length

        # Expand around the center as long as the characters match
        while left >= 0 and right < n and string[left] == string[right]:
            left -= 1
            right += 1

        # Calculate the length of the palindrome substring
        length = right - left - 1

        # Update the start and max_length if a longer palindrome is found
        if length > max_length:
            start = left + 1
            max_length = length

    # Iterate through each character as the potential center of the palindrome
    for i in range(n):
        # For odd-length palindromes
        expand_around_center(i, i)

        # For even-length palindromes
        expand_around_center(i, i + 1)

    # Return the longest palindrome substring
    return string[start:start + max_length]
```

```python
string = "babad"
longest_palindrome = longest_palindrome_substring(string)
print(longest_palindrome)
```

44. Implement a function to check if a given number is a perfect number.

```python
def is_perfect_number(number):
    if number <= 0:
        return False

    divisors = [1]  # List to store the divisors

    # Find divisors of the number
    for i in range(2, int(number ** 0.5) + 1):
        if number % i == 0:
            divisors.append(i)
            if i != number // i:
                divisors.append(number // i)

    # Check if the sum of divisors equals the number
    return sum(divisors) == number
```

```python
number1 = 6
print(is_perfect_number(number1))

number2 = 28
print(is_perfect_number(number2))

number3 = 12
print(is_perfect_number(number3))
```
